```python
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt
from matplotlib import patches
```

```python
A = np.array([[3, 0], [0, -2]])
B = np.array([[2, 0], [0, 3]])
C = np.array([[1, 2], [0, 2]])
D = np.array([[1, 1], [0, 0]])
E = np.array([[1, 1], [1, 1]])
```

```python
def LSV(X):
    XXT = X@X.T
    lam, v = np.linalg.eig(XXT)
    sig = np.sqrt(lam)
    return v, sig

def USV(X, v):
    sig_u = X@v
    return sig_u
```

```python
v_1, sig_1 = LSV(A)
u_1 = USV(A, v_1)

fig_1, ax_1 = plt.subplots()
circ= plt.Circle((0, 0), 1/2.17, fill=False)
origin = np.array([[0, 0],[0, 0]])
plt.quiver(*origin, v_1[:, 0], v_1[:, 1], color = ["#5C6BC0", "#009688"], scale = 2.17)
ax_1.add_patch(circ)
plt.axis('off')
ax_1.set_aspect('equal')
plt.show()
```

```python
fig_2, ax_2 = plt.subplots()
Ellipse = patches.Ellipse((0, 0), sig_1[0], sig_1[1], fill=False)
origin = np.array([[0, 0],[0, 0]])
Q = plt.quiver(*origin, u_1[:, 0], u_1[:, 1], color = ["#5C6BC0", "#009688"], scale = 6.6)
ax_2.add_patch(Ellipse)
plt.axis('off')
ax.set_aspect('equal')
plt.show()
```
