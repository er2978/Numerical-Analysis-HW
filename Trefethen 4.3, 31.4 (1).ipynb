```python
A = np.array([[3, 0], [0, -2]])
B = np.array([[2, 0], [0, 3]])
C = np.array([[1, 2], [0, 2]])
D = np.array([[1, 1], [0, 0]])
E = np.array([[1, 1], [1, 1]])
```

```python
def SVD(X):
    U, s, VT = np.linalg.svd(X)
    V= VT.T
    Sig = np.diag(s) 
    return U, Sig, V
```

```python
def AutoSmallSig(n):
    X = np.zeros((n, n))
    for i in range(n):
        for j in range(n):
            X[i, j]= 0 
            if j > i:
                X[i, j] = 1
            if j == i:
                X[i,j] = 0.1
    U_up, Sig_up, V_up = SVD(X)
    return Sig_up[n-1, n-1]
print(AutoSmallSig(7))
```

```python
def ManuSmallSig(n):
    X = np.zeros((n, n))
    for i in range(n):
        for j in range(n):
            X[i, j]= 0 
            if j > i:
                X[i, j] = 1
            if j == i:
                X[i,j] = 0.1
    lambs = sorted(np.linalg.eigvals(X.T@X), reverse = True)
    Cigs = np.sqrt(np.abs(lambs))
    return Cigs[n-1]
print(ManuSmallSig(7))
```

### Note:
np.linalg.eigvals is registering some values as NaN so they are omitted from the square root after a certain m value (m = 7)

```python
m_val = np.arange(1, 30)

def t_arr(m_arr):
    ends = np.zeros(len(m_arr))
    end_s = np.zeros(len(m_arr))
    
    for i in range(len(m_arr)):
        m = m_arr[i]
        
        start = time.time()
        sig1 = AutoSmallSig(m)
        ends[i] = time.time() - start
        
        start_ = time.time()
        sig_1 = ManuSmallSig(m)
        end_s[i] = time.time() - start_
    return ends, end_s
ends, end_s = t_arr(m_val)

fig, ax = plt.subplots(1,2,sharex = True, sharey = True, figsize = (10,10))
plt.xlabel('m value', fontsize = 16)
plt.ylabel('Time', fontsize = 16)
plt.title('loglog plots of m value vs time to get smallest singular value')
ax[0].loglog(m_val, ends, c = '#6D21FF')
ax[0].set_title('Automatic Singular Value Runtime')
ax[1].loglog(m_val, end_s, c = '#6D21FF')
ax[1].set_title('Manual Singular Value Runtime')
plt.show()
```
