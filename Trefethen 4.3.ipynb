```python
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt
```

```python
A = np.array([[3, 0], [0, -2]])
B = np.array([[2, 0], [0, 3]])
C = np.array([[1, 2], [0, 2]])
D = np.array([[1, 1], [0, 0]])
E = np.array([[1, 1], [1, 1]])
```

```python
def LSV(X):
    XXT = X@X.T
    lam, v = np.linalg.eig(XXT)
    sig = np.sqrt(lam)
    return v, sig
v_1, sig_1 = LSV(A)
```

```python
v_1, sig_1 = LSV(A)

fig, ax = plt.subplots()
figsize = (5,5)
circ= plt.Circle((0, 0), 1/2.15, fill=False)
origin = np.array([[0, 0],[0, 0]])
plt.quiver(*origin, v_1[:, 0], v_1[:, 1], color = ["#5C6BC0", "#009688"], scale = 2.15)
ax.add_patch(circ)
ax.set_aspect('equal', adjustable='box')
plt.show()
```
