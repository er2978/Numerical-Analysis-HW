```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def func(x, y):
    f_x = y - x**2
    return f_x
```

```python
def Babylonian(x, y, n):
    xnlist = [x]
    xn = x
    error = [1]
    for i in range(n):
        xn = 0.5*(xn+(y/xn))
        xnlist.append(xn)
        error.append(abs(xn-np.sqrt(y))/abs(x-np.sqrt(y)))
        i+=1
        ops = 3*i
        if func(xn, y) == 0:
            break
    return [xnlist, y, i, ops, error]
```

```python
def Secant(a, b, y, n):
    xnlist = [a, b]
    error = [1]
    for i in range(n):
        xn = xnlist[i+1] - ((xnlist[i+1]**2 - y)/(xnlist[i+1]+xnlist[i]))
        xnlist.append(xn)
        error.append(abs(xn-np.sqrt(y))/abs(xnlist[0]-np.sqrt(y)))
        i+=1
        ops = 5*i
        if func(xn, y) == 0:
            break
    return [xnlist, y, i, ops, error]
```

```python
def Bisect(a, b, y, n):
    i = 0
    x0 = a
    x1 = b
    xnlist = [x0, x1]
    error = [1]
    xn = 0.5*(x0+x1)
    xnlist.append(xn)
    while func(xn, y)*func(x0, y) < 0:
        x1 = xn
        xn = 0.5*(x0 + x1)
        xnlist.append(xn)
    while func(xn, y)*func(x0, y) > 0:
        x0 = xn
        xn = 0.5*(x0 + x1)
        xnlist.append(xn)
    i = len(xnlist)
    ops = 2*(len(xnlist) - 2)
    for j in range(len(xnlist)):
        error.append(abs(xnlist[j]-np.sqrt(y))/abs(xnlist[2]-np.sqrt(y)))
    return [xnlist, y, i, error, ops]
```

```python
x = 3
y = 16
a = 1
b = 20
n = 100
```

```python
print("Babylonian Iterations: ", Babylonian(x, y, n)[0]) 
print("starting point:", x) 
print("y=", y)
print("n=", Babylonian(x, y, n)[2])
print("floating point operations:", Babylonian(x, y, n)[3])
print("relative error:", Babylonian(x, y, n)[4])
```

```python
print("Secant Iterations: ", Secant(a, b, y, n)[0]) 
print("interval:", a, b)
print("y=", y)
print("n=", Secant(a, b, y, n)[2])
print("floating point operations:", Secant(a, b, y, n)[3])
print("relative error:", Secant(a, b, y, n)[4])
```

```python
print("Bisection Iterations: ", Bisect(a, b, y, n)[0]) 
print("interval:", a, b) 
print("y=", y)
print("n=", Secant(a, b, y, n)[2])
print("floating point operations:", Bisect(a, b, y, n)[3])
print("relative error:", Bisect(a, b, y, n)[4])
```

## 4.3 
In general I found Bisection to be both slow and inaccurate, but by golly is it cheap -- it only costs 2 floating point operations per iteration. I found the Babylonian method most reliable, though it is the most expensive (being a Newton iteration.) They all seem to misbehave when negative numbers are involved, churning out negative values for $x^*$ though that's to be expected since the negative value is also a root to f(x, y). Secant and Bisection work best on a smaller interval, whereas Babylonian doesn't really care.
